# Portfolio Optimization: Analysis on TSLA, BND, and SPY Stock Data

## Introduction

In this notebook, we explore the process of portfolio optimization using the historical stock data for TSLA, BND, and SPY. Portfolio optimization is a method used to select the best mix of assets to maximize returns while minimizing risk. By applying advanced statistical models and optimization techniques, we aim to construct a well-balanced portfolio that meets specified financial goals.

### Objectives
- **Data Collection and Preprocessing**: Load, clean, and combine the historical and Forecasted stock data for TSLA, BND, and SPY.
- **Annualized Returns Calculation**: Compute the annualized returns for each stock.
- **Portfolio Statistics**: Calculate key portfolio metrics such as return, risk, and Sharpe Ratio.
- **Trend Analysis**: Analyze forecasted data to identify trends and patterns.
- **Optimization**: Apply optimization techniques to determine the optimal portfolio allocation.
- **Risk and Opportunity Assessment**: Evaluate the potential risks and opportunities associated with the optimized portfolio.

### Dataset
The dataset used in this analysis consists of historical stock prices for TSLA, BND, and SPY and Forecasted data using LSTM Model for the next 12 months considering 21 stock days per month. The data includes daily closing prices, which will be used to calculate returns, risk, and perform the optimization.
